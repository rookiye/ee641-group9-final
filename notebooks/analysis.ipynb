# EE641 Final Project - Analysis Notebook

This notebook analyzes the results from COLD and GCG experiments.

```python
import wandb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize W&B API
api = wandb.Api()
runs = api.runs("vthiyaga-usc/EE641 project")
```

## Collect Results

```python
# Collect data from W&B runs
data = []
for run in runs:
    if run.state == 'finished':
        data.append({
            'method': run.config.get('method', 'Unknown'),
            'catalog': run.config.get('catalog', ''),
            'product_idx': run.config.get('target_product_idx', -1),
            'final_rank': run.summary.get('eval/product_rank', -1),
            'success': run.summary.get('success', False)
        })

df = pd.DataFrame(data)
df.head()
```

## Method Comparison

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='method', y='final_rank')
plt.title('COLD vs GCG Performance', fontsize=14, fontweight='bold')
plt.ylabel('Product Rank (lower is better)')
plt.savefig('figures/method_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Success Rate

```python
success_rate = df.groupby('method')['success'].mean() * 100
print("Attack Success Rate (ASR):")
print(success_rate)
```
